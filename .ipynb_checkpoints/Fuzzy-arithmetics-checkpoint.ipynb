<h1 style="color:slateblue;font-family:courier;">Fuzzy Arithmetics Operations</h1>
<h2 style="color:slateblue;font-family:courier;">Interval arithmetic</h2>

<ul style="font-size:20px;font-family:courier;">
    <li>Four fuzzy interval operations are implemented in the <i>fuzzy-arithmetic.R</i> script: 
      <ul>
          <li>Addition</li>
          <li>Subtraction</li>
          <li>Multiplication</li>
          <li>Division</li>
      </ul>
    </li>  
</ul> 

```r
source("fuzzy-arithmetic.R")
```

<h3 style="color:slateblue;font-family:courier;">Defining the triangular membership function by given the three points, minimum, mean and maximum values for each fuzzy set</h3>

```r
fuzzy.set.a = c(min.value = 1, mean.value = 4, max.value = 6)
fuzzy.set.b = c(min.value = 3, mean.value = 5, max.value = 7)
```

<h3 style="color:slateblue;font-family:courier;">The <i>plot.ops</i> function gives the plot of all arithmetic operations </h3>

```r
plot.ops(fuzzy.set.a, fuzzy.set.b)
```

<h3 style="color:slateblue;font-family:courier;"><i>plot.ops</i> gives the individual plot of each operation from its name (third parameter) </h3>

```r
plot.ops(fuzzy.set.a, fuzzy.set.b, "add")
plot.ops(fuzzy.set.a, fuzzy.set.b, "sub")
```

```r
plot.ops(fuzzy.set.a, fuzzy.set.b, "mult")
plot.ops(fuzzy.set.a, fuzzy.set.b, "div")
```
